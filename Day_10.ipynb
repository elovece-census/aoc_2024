## Part 1

```python
from collections import deque
```

```python
# Convert map into 2D list of integers
grid = [list(map(int, row)) for row in lines]
for line in grid: print(line)
```

```python
rows, cols = len(grid), len(grid[0])
```

```python
# Directions for moving up, down, left, and right
directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
```

```python
total_score = 0
total_visited_nines = set()
```

```python
for r in range(rows):
    for c in range(cols):
        
        if grid[r][c] == 0:
            
            start = (r, c)            
            queue = deque([start])
            local_visited = set([start])
            reachable_nines = set()
            
            while queue:
                x, y = queue.popleft()
                current_height = grid[x][y]
                
                # Explore neighbors
                for dx, dy in directions:
                    nx, ny = x + dx, y + dy
                    
                    if 0 <= nx < rows and 0 <= ny < cols:
                        # Check if height increases by 1
                        if (nx, ny) not in local_visited and grid[nx][ny] == current_height + 1:
                            local_visited.add((nx, ny))
                            queue.append((nx, ny))
                            
                        # Record all 9s that are reached
                        if grid[nx][ny] == 9 and (nx, ny) not in reachable_nines:
                            reachable_nines.add((nx, ny))
                            
            total_visited_nines.update(reachable_nines)       
            total_score += len(reachable_nines)
                            
total_score
```

```python
from collections import deque

# Sample topographic map (replace this with the 57x57 input as needed)
# Each row is represented as a string of digits.
topographic_map = [
    "89010123",
    "78121874",
    "87430965",
    "96549874",
    "45678903",
    "32019012",
    "01329801",
    "10456732",
]

# Convert the map into a 2D list of integers
grid = [list(map(int, row)) for row in topographic_map]
rows, cols = len(grid), len(grid[0])

# Directions for moving up, down, left, and right
directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def bfs_trailhead_score_with_global_tracking(start, global_visited_nines):
    """Perform BFS from a trailhead and calculate its score, ensuring no overcounting."""
    queue = deque([start])
    local_visited = set([start])  # Visited cells in this BFS
    reachable_nines = set()

    while queue:
        x, y = queue.popleft()
        current_height = grid[x][y]

        # Explore neighbors
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < rows and 0 <= ny < cols:
                # Check if the height increases exactly by 1
                if (nx, ny) not in local_visited and grid[nx][ny] == current_height + 1:
                    local_visited.add((nx, ny))
                    queue.append((nx, ny))
                
                # If the height is 9 and not globally visited, add it to reachable_nines
                if grid[nx][ny] == 9 and (nx, ny) not in reachable_nines:
                    reachable_nines.add((nx, ny))

    # Mark all reachable 9s as globally visited to prevent double-counting
    new_nines = reachable_nines - global_visited_nines
    global_visited_nines.update(new_nines)

    return len(reachable_nines)


# Calculate total score with corrected logic
total_score_corrected = 0
global_visited_nines = set()  # Global tracking of visited 9s

for r in range(rows):
    for c in range(cols):
        if grid[r][c] == 0:  # Trailhead found
            total_score_corrected += bfs_trailhead_score_with_global_tracking((r, c), global_visited_nines)

total_score_corrected
```
