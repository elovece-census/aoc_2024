## Part 2

```python
# Parse the dense disk map into a block-by-block representation.
blocks_large = []
file_id = 0
is_file = True  # First char is a file, then alternates bet. free space and file
for char in disk_map:
    length = int(char)
    
    if is_file:
        blocks_large.extend([str(file_id)] * length)
        file_id += 1
    else:
        blocks_large.extend(["."] * length)
        
    is_file = not is_file
```

```python
# Move whole files into the leftmost space that can fit them entirely
# in order of decreasing file ID

# Identify files and their positions
file_positions = {}
for i, block in enumerate(blocks_large):
    if block != ".":
        file_positions.setdefault(block, []).append(i)

# Process files in descending order of file ID
for file_id in sorted(file_positions.keys(), key=int, reverse=True):
    file_positions[file_id].sort()  # Ensure positions are in order
    file_length = len(file_positions[file_id])

    # Find spans of free space
    free_spans = []
    start = None
    for i, block in enumerate(blocks_large):
        
        if block == "." and start is None:
            start = i  # Start a new span
        elif block != "." and start is not None:
            free_spans.append((start, i - start))
            start = None
    
    if start is not None:
        free_spans.append((start, len(blocks_large) - start))

    # Look for a span large enough to fit the file
    for start, length in free_spans:
        if length >= file_length:
            
            # Move the file to this span
            for pos in file_positions[file_id]:
                blocks_large[pos] = "."  # Clear the old file positions
                
            for i in range(file_length):
                blocks_large[start + i] = file_id  # Place the file
                
            break  # File has been moved, stop processing it
```

```python
def find_free_space(blocks, length):
    free_start = None  # Start index of the current free space span
    free_length = 0
    for i, block in enumerate(blocks):
        
        if block == ".":
            
            if free_start is None:
                free_start = i  # Start a new span
                
            free_length += 1
            
        if free_length >= length:
            return free_start
        else:
            free_start = None
            free_length = 0
        
    return None
```

```python
def move_largest_file_to_free_space(blocks):
    """
    Moves the largest set of consecutive file numbers that can fit into
    the first span of free space (dots). Replaces the file's original position with dots.
    """
    #blocks = list(blocks)
    
   # Step 1: Find the largest file that fits into the free space span
    max_file_id = int(max(blocks))
    for digit in range(max_file_id, -1, -1):
        digit_str = str(digit)

        visited_digits = set()
        i = 0
        while i < len(blocks):
            if blocks[i] == digit_str:
                
                if digit_str in visited_digits:
                    i += 1
                    continue
                else:
                
                    file_start = i
                    file_length = 0

                    # Count the length of the current file
                    while i < len(blocks) and blocks[i] == digit_str:
                        file_length += 1
                        i += 1

                    # Check if this file can fit and is the largest seen so far
                    free_start = find_free_space(blocks, file_length)
                    if free_start is not None:
                        # Clear the original file position
                        for j in range(file_start, file_start + file_length):
                            blocks[j] = "."

                        # Place the file in the free space
                        for j in range(free_start, free_start + file_length):
                            blocks[j] = digit_str

                    #break
                    visited_digits.add(digit_str)
            else:
                i += 1

    return blocks
```
